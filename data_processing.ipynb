```python
from mlxtend.data import loadlocal_mnist
import platform
```

```python
def load_data(train,lable):
    X, y = loadlocal_mnist(
              images_path=train, 
              labels_path=lable)
    return X,y
```

```python
X_train,y_train = load_data('datasets/train-images.idx3-ubyte','datasets/train-labels.idx1-ubyte')
X_test,y_test = load_data('datasets/t10k-images.idx3-ubyte','datasets/t10k-labels.idx1-ubyte')
```

```python
X_train = np.float32(X_train)
y_train = np.int32(y_train)
X_test = np.float32(X_test)
y_test = np.int32(y_test)
```

```python
X_train = X_train.reshape(60000,1,28,28)
X_test = X_test.reshape(10000,1,28,28)
```

Split train data & val data

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42)
```

Save data

```python
torch.save(X_train,'datasets/X_train')
torch.save(X_test,'datasets/X_test')
torch.save(y_train,'datasets/y_train')
torch.save(y_test,'datasets/y_test')
torch.save(X_val,'datasets/X_val')
torch.save(y_val,'datasets/y_val')
```
